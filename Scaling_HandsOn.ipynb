```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
df.columns
```

```python
ss.fit_transform(df[['speeding', 'alcohol', 'not_distracted', 'no_previous',
       'ins_premium', 'ins_losses']])
```

```python
df_ss = ss.fit_transform(df[['speeding', 'alcohol', 'not_distracted', 'no_previous',
       'ins_premium', 'ins_losses']])
df_ss.shape
```

```python
pd.DataFrame(df_ss,columns=['speeding', 'alcohol', 'not_distracted', 'no_previous',
       'ins_premium', 'ins_losses'])
```
